```python
task4_f1.round(4),task4_f2.round(4)
```

```python
task4_w1 = 2*np.pi*task4_f1
task4_w2 = 2*np.pi*task4_f2
task4_w1, task4_w2
```

```python
task4_f0 = task4_f1*task4_f2*np.sqrt(2)/np.sqrt(task4_f1**2+task4_f2**2)
task4_w0 = 2*np.pi*task4_f0
task4_f0_avg = np.average(task4_f0)
task4_w0_avg = 2*np.pi*task4_f0_avg
task4_f0_avg
```

```python
M1 = L*(task4_w0_avg**2/task4_w1**2 - 1)
M2 = -L*(task4_w0_avg**2/task4_w2**2 - 1)
M = (M1+M2)/2
K = L*M

M.round(4)
```

```python
def f4(x, n, A):
  return A*x**n
```

```python
popt4m, pcov4m = curve_fit(f4, task4_cm, M, p0 = [-1,4])
popt4k, pcov4k = curve_fit(f4, task4_cm, K, p0 = [-1,4])
popt4m, np.sqrt(pcov4m[0][0])
```

```python
x = np.linspace(4,15,1000)
plt.scatter(task4_cm, M, label = 'data')
plt.plot(x,f4(x, *popt4m), label = 'fit', c = 'r')
plt.xlabel('Distance between coils (cm)')
plt.ylabel('Mutual Inductance M (H)')
plt.legend()
plt.savefig('task42.png', dpi = 500)
```
